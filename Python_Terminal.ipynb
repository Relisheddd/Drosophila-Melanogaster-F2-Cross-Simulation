Question:  Do probabilistic simulations of Drosophila eye color inheritance better support an X-linked model or an autosomal model?

Assuming White Eyes is a Sex Linked Trait


```python
import random

#
#ASSUMING WHITE EYES IS SEX LINKED TRAIT
#

print("Assuming White Eyes Is A Sex Linked Trait-")
print("")

RSexX = 0
DSexX = 1
SexY = 10

FemaleP = [DSexX, DSexX]
MaleP = [RSexX, SexY]

def printPhGeno(fly):
  if(fly[0] == 10 or fly[1] == 10):
    if((fly[0]+fly[1])==10):
      return "fly is male, pheno is white eyes, geno (X^a, Y)"
    else:
      return "fly is male, pheno is red eyes, geno (X^A, Y)"
  else:
    if((fly[0]+fly[1])==0):
      return "fly is female, pheno is white eyes, geno is (X^a, X^a)"
    else:
      if((fly[0]+fly[1])==1):
        return "fly is female, pheno is red eyes, geno is (X^A, X^a)"
      elif((fly[0]+fly[1])==2):
        return "fly is female, pheno is red eyes, geno is (X^A, X^A)"

def tallyPhGeno(fly):
  global Mwhite, Mred, Fwhite, FhomoRed, FheteroRed
  if(fly[0] == 10 or fly[1] == 10):
    if((fly[0]+fly[1])==10):
      Mwhite += 1
    else:
      Mred += 1
  else:
    if((fly[0]+fly[1])==0):
      Fwhite += 1
    else:
      if((fly[0]+fly[1])==1):
        FheteroRed += 1
      elif((fly[0]+fly[1])==2):
        FhomoRed += 1

def tallyPhGeno2(fly):
  global Mwhite2, Mred2, Fwhite2, FhomoRed2, FheteroRed2
  if(fly[0] == 10 or fly[1] == 10):
    if((fly[0]+fly[1])==10):
      Mwhite2 += 1
    else:
      Mred2 += 1
  else:
    if((fly[0]+fly[1])==0):
      Fwhite2 += 1
    else:
      if((fly[0]+fly[1])==1):
        FheteroRed2 += 1
      elif((fly[0]+fly[1])==2):
        FhomoRed2 += 1


def crossTwo(mom, dad):
    gene_from_mom = random.choice(mom)
    gene_from_dad = random.choice(dad)
    return [gene_from_mom, gene_from_dad]

print("Parental Data:")
print("Mom: " + printPhGeno(FemaleP))
print("Dad: " + printPhGeno(MaleP))

#F1 generation
F1 = []
Mwhite = 0
Mred = 0
Fwhite = 0
FhomoRed = 0
FheteroRed = 0

for i in range(1000):
  F1.append(crossTwo(FemaleP, MaleP))

for i in F1:
  tallyPhGeno(i)

# Update F1T after all tallies are done
F1T = [Mwhite, Mred, Fwhite, FhomoRed, FheteroRed]

F1_males = []
F1_females = []

for fly in F1:
    if (fly[0] == 10 or fly[1] == 10):  # has Y chromosome
        F1_males.append(fly)
    else:
        F1_females.append(fly)

print("")
print("F1 Results:")
print("White males:", F1T[0])
print("Red males:", F1T[1])
print("White females:", F1T[2])
print("Homozygous red females:", F1T[3])
print("Heterozygous red females:", F1T[4])

#F2 generation
F2 = []
Mwhite2 = 0
Mred2 = 0
Fwhite2 = 0
FhomoRed2 = 0
FheteroRed2 = 0

for i in range(1000):
  # Randomly select a female and a male from F1 to be parents for F2
  chosen_mom_genotype = random.choice(F1_females)
  chosen_dad_genotype = random.choice(F1_males)
  F2.append(crossTwo(chosen_mom_genotype, chosen_dad_genotype))

for i in F2:
  tallyPhGeno2(i)

F2T = [Mwhite2, Mred2, Fwhite2, FhomoRed2, FheteroRed2]

print("")
print("F2 Results:")
print("White males:", F2T[0])
print("Red males:", F2T[1])
print("White females:", F2T[2])
print("Homozygous red females:", F2T[3])
print("Heterozygous red females:", F2T[4])

#
#ASSUMING WHITE EYES IS AUTOSOMAL TRAIT
#

print("")
print("Assuming White Eyes Is An Autosomal Trait-")
print("")

Rec = 0
Dom = 1

Pfemale = [Dom, Dom]
Pmale = [Rec, Rec]

def printPhGeno(fly):
  if((fly[0]+fly[1])==0):
    return "pheno is white eyes, geno is (a, a)"
  else:
    if((fly[0]+fly[1])==1):
      return "pheno is red eyes, geno is (A, a)"
    elif((fly[0]+fly[1])==2):
      return "pheno is red eyes, geno is (A, A)"

def MtallyPhGeno(fly):
  global Mwhite, MhomoRed, MheteroRed
  if((fly[0]+fly[1])==0):
    Mwhite += 1
  else:
    if((fly[0]+fly[1])==1):
      MheteroRed += 1
    elif((fly[0]+fly[1])==2):
      MhomoRed += 1

def FtallyPhGeno(fly):
  global Fwhite, FhomoRed, FheteroRed
  if((fly[0]+fly[1])==0):
    Fwhite += 1
  else:
    if((fly[0]+fly[1])==1):
      FheteroRed += 1
    elif((fly[0]+fly[1])==2):
      FhomoRed += 1

def MtallyPhGeno2(fly):
  global Mwhite2, MhomoRed2, MheteroRed2
  if((fly[0]+fly[1])==0):
    Mwhite2 += 1
  else:
    if((fly[0]+fly[1])==1):
      MheteroRed2 += 1
    elif((fly[0]+fly[1])==2):
      MhomoRed2 += 1

def FtallyPhGeno2(fly):
  global Fwhite2, FhomoRed2, FheteroRed2
  if((fly[0]+fly[1])==0):
    Fwhite2 += 1
  else:
    if((fly[0]+fly[1])==1):
      FheteroRed2 += 1
    elif((fly[0]+fly[1])==2):
      FhomoRed2 += 1

def crossTwo(mom, dad):
  gene_from_mom = random.choice(mom)
  gene_from_dad = random.choice(dad)
  return [gene_from_mom, gene_from_dad]

print("Parental Data:")
print("Mom: " + printPhGeno(Pfemale)) # Changed FemaleP to Pfemale
print("Dad: " + printPhGeno(Pmale))
print("")

#F1 generation
F1_males = []
F1_females = []
Mwhite = 0
MhomoRed = 0
MheteroRed = 0
Fwhite = 0
FhomoRed = 0
FheteroRed = 0

for i in range(1000):
  if(random.randint(1,2)==1):
    F1_males.append(crossTwo(Pfemale, Pmale))
  else:
    F1_females.append(crossTwo(Pfemale, Pmale))

for i in F1_males:
  MtallyPhGeno(i)

for i in F1_females:
  FtallyPhGeno(i)

# Update F1T after all tallies are done
F1T = [Mwhite, MhomoRed, MheteroRed, Fwhite, FhomoRed, FheteroRed]

print("F1 Results:")
print("White males:", F1T[0])
print("Homozygous red males:", F1T[1])
print("Heterozygous red males:", F1T[2])
print("White females:", F1T[3])
print("Homozygous red females:", F1T[4])
print("Heterozygous red females:", F1T[5])
print("")

#F2 generation
F2_males = []
F2_females = []
Mwhite2 = 0
MhomoRed2 = 0
MheteroRed2 = 0
Fwhite2 = 0
FhomoRed2 = 0
FheteroRed2 = 0

for i in range(1000):
  # Randomly select a female and a male from F1 to be parents for F2
  if(random.randint(1,2)==1):
    chosen_mom_genotype = random.choice(F1_females)
    chosen_dad_genotype = random.choice(F1_males)
    F2_females.append(crossTwo(chosen_mom_genotype, chosen_dad_genotype))
  else:
    chosen_mom_genotype = random.choice(F1_females)
    chosen_dad_genotype = random.choice(F1_males)
    F2_males.append(crossTwo(chosen_mom_genotype, chosen_dad_genotype))

for i in F2_males:
  MtallyPhGeno2(i)

for i in F2_females:
  FtallyPhGeno2(i)

# Update F1T after all tallies are done
F2T = [Mwhite2, MhomoRed2, MheteroRed2, Fwhite2, FhomoRed2, FheteroRed2]

print("F1 Results:")
print("White males:", F2T[0])
print("Homozygous red males:", F2T[1])
print("Heterozygous red males:", F2T[2])
print("White females:", F2T[3])
print("Homozygous red females:", F2T[4])
print("Heterozygous red females:", F2T[5])
```

```python
import random

# =========================
# SETTINGS (EDIT THESE)
# =========================
N = 5000          # offspring per generation
RUNS = 10         # change to 100 if you want

# =========================
# MODEL 1: SEX-LINKED (X-linked recessive)
# =========================
def run_sexlinked(N=5000):
    # Alleles:
    # X^a (white) = 0
    # X^A (red)   = 1
    # Y           = 10
    X_a = 0
    X_A = 1
    Y = 10

    # Parental cross (Morgan-style):
    # Female: X^A X^A (red)
    # Male:   X^a Y   (white)
    FemaleP = [X_A, X_A]
    MaleP = [X_a, Y]

    def crossTwo(mom, dad):
        return [random.choice(mom), random.choice(dad)]

    def is_male(fly):
        return (fly[0] == Y or fly[1] == Y)

    # -------------------------
    # F1 generation
    # -------------------------
    F1 = [crossTwo(FemaleP, MaleP) for _ in range(N)]

    F1_males = [fly for fly in F1 if is_male(fly)]
    F1_females = [fly for fly in F1 if not is_male(fly)]

    # -------------------------
    # F2 generation
    # -------------------------
    F2 = []
    for _ in range(N):
        mom = random.choice(F1_females)
        dad = random.choice(F1_males)
        F2.append(crossTwo(mom, dad))

    # -------------------------
    # Tally F2 phenotypes
    # -------------------------
    # We only need phenotype totals for the report:
    # Mwhite, Mred, Fwhite, Fred
    Mwhite = 0
    Mred = 0
    Fwhite = 0
    Fred = 0

    for fly in F2:
        if is_male(fly):
            # male: X + Y
            if (fly[0] + fly[1]) == Y:   # X^a + Y = 0 + 10 = 10
                Mwhite += 1
            else:
                Mred += 1
        else:
            # female: X + X
            # white only if X^a X^a = 0
            if (fly[0] + fly[1]) == 0:
                Fwhite += 1
            else:
                Fred += 1

    return [Mwhite, Mred, Fwhite, Fred]


# =========================
# MODEL 2: AUTOSOMAL RECESSIVE
# =========================
def run_autosomal(N=5000):
    # Alleles:
    # a (white) = 0
    # A (red)   = 1
    a = 0
    A = 1

    # Parental cross:
    # Female: AA (red)
    # Male:   aa (white)
    Pfemale = [A, A]
    Pmale = [a, a]

    def crossTwo(mom, dad):
        return [random.choice(mom), random.choice(dad)]

    # -------------------------
    # F1 generation
    # -------------------------
    # In autosomal, sex doesn't affect inheritance,
    # but we still assign sex randomly for comparison.
    F1 = [crossTwo(Pfemale, Pmale) for _ in range(N)]

    # Randomly assign sexes 50/50
    F1_males = []
    F1_females = []
    for fly in F1:
        if random.randint(1, 2) == 1:
            F1_males.append(fly)
        else:
            F1_females.append(fly)

    # -------------------------
    # F2 generation
    # -------------------------
    F2 = []
    for _ in range(N):
        mom = random.choice(F1_females)
        dad = random.choice(F1_males)
        F2.append(crossTwo(mom, dad))

    # Randomly assign sexes again for F2
    F2_males = []
    F2_females = []
    for fly in F2:
        if random.randint(1, 2) == 1:
            F2_males.append(fly)
        else:
            F2_females.append(fly)

    # -------------------------
    # Tally F2 phenotypes
    # -------------------------
    # phenotype: white only if aa (0)
    Mwhite = sum(1 for fly in F2_males if (fly[0] + fly[1]) == 0)
    Mred   = len(F2_males) - Mwhite

    Fwhite = sum(1 for fly in F2_females if (fly[0] + fly[1]) == 0)
    Fred   = len(F2_females) - Fwhite

    return [Mwhite, Mred, Fwhite, Fred]


# =========================
# RUN MULTIPLE TIMES + AVERAGE
# =========================
def summarize(results_list):
    # results_list is a list of [Mwhite, Mred, Fwhite, Fred]
    cols = list(zip(*results_list))  # transpose

    avg = [sum(col) / len(col) for col in cols]
    mn = [min(col) for col in cols]
    mx = [max(col) for col in cols]

    return avg, mn, mx


sexlinked_runs = []
autosomal_runs = []

for _ in range(RUNS):
    sexlinked_runs.append(run_sexlinked(N))
    autosomal_runs.append(run_autosomal(N))

sex_avg, sex_min, sex_max = summarize(sexlinked_runs)
auto_avg, auto_min, auto_max = summarize(autosomal_runs)

# =========================
# PRINT RESULTS
# =========================
labels = ["White Males", "Red Males", "White Females", "Red Females"]

print("\n==============================")
print("SIMULATION SUMMARY")
print("==============================")
print("Offspring per generation (N):", N)
print("Number of runs:", RUNS)

print("\n--- SEX-LINKED MODEL (X-linked recessive) ---")
for i in range(4):
    print(f"{labels[i]} | avg = {sex_avg[i]:.1f} | min = {sex_min[i]} | max = {sex_max[i]}")

print("\n--- AUTOSOMAL MODEL (recessive) ---")
for i in range(4):
    print(f"{labels[i]} | avg = {auto_avg[i]:.1f} | min = {auto_min[i]} | max = {auto_max[i]}")


```
